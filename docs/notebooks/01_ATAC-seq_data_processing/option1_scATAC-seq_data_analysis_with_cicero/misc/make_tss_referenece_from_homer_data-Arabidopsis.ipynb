# 2. Make tss bed file

```python
def load_tss_data(ref_genome):

    tss = pd.read_csv(f"data/genomes/{ref_genome}/{ref_genome}.basic.annotation",
                      header=None, delimiter="\t")
    tss = tss[tss[5] == "P"]
    
    #return(tss)
    tss.columns = ["annotation", "Chr", "Start", "End", "Strand", "5", "6"]
    tss["gene_id_full"] = [i.replace("promoter-TSS (", "").replace(")", "") for i in tss["annotation"]]   
    tss["gene_id"] = [i.split(".")[0] for i in tss["gene_id_full"]]
    tss["Distance to TSS"] = np.nan
    
    tss = tss[["Chr", "Start", "End", "gene_id", "Distance to TSS", "Strand"]]
    
    
    idx = tss.index[tss.Start < 0]
    tss.loc[idx, "Start"] = 0

    return tss



tss = load_tss_data(ref_genome=ref_genome)
```
